```python
# librariessss

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

```python
train_df = pd.read_csv('train.csv')
```

```python
train_df.head()
```

```python
pd.set_option('display.max_rows', None)
print(train_df.dtypes)
```

```python
int64_df = pd.DataFrame(train_df.select_dtypes(include=['int64']))
pd.set_option('display.max_columns', None)
int64_df.head()
```

```python
int64_df.info()
```

```python
clean_df = int64_df.drop(columns = ['Id', 'MSSubClass', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd','YrSold'])
```

```python
clean_df.head()
```

### Problem 2 : Statistical Summary
Now that you have a “clean” DataFrame with only Numeric variables, we can safely perform standard statistics.
> Find the Summary Statistics (Mean, Median, Quartiles etc.) of SalePrice from the Numeric DataFrame.

> Visualize the summary statistics and distribution of SalePrice using standard Box-Plot, Histogram, KDE.

> Find the Summary Statistics (Mean, Median, Quartiles etc) of LotArea from the Numeric DataFrame.

> Visualize the summary statistics and distribution of LotArea using standard Box-Plot, Histogram, KDE.

> Plot SalePrice (y-axis) vs LotArea (x-axis) using jointplot and find the Correlation between the two.

```python
SalePrice = pd.DataFrame(clean_df['SalePrice'])
SalePrice.describe()
```

```python
f = plt.figure(figsize=(16,8))
sb.violinplot(data = SalePrice, orient = "h")
```

```python
f = plt.figure(figsize=(16,8))
sb.boxplot(data = SalePrice, orient = "h")
```

```python
f = plt.figure(figsize=(16,8))
sb.kdeplot(data = SalePrice)
```

```python
f = plt.figure(figsize=(16,8))
sb.histplot(data = SalePrice)
```

```python
LotArea = pd.DataFrame(clean_df['LotArea'])
```

```python
LotArea.describe()
```

```python
f = plt.figure(figsize=(16,8))
sb.violinplot(data = LotArea, orient = 'h')
```

```python
f = plt.figure(figsize=(16,8))
sb.boxplot(data = LotArea, orient = 'h')
```

```python
f = plt.figure(figsize=(16,8))
sb.kdeplot(data = LotArea)
```

```python
f = plt.figure(figsize=(16,8))
sb.histplot(data = LotArea)
```

```python
jointDF = pd.concat([SalePrice, LotArea], axis = 1).reindex(SalePrice.index)
jointDF
```

```python
sb.jointplot(data = jointDF, x = "SalePrice", y = "LotArea", height = 12)
```

```python
jointDF.corr()
```
